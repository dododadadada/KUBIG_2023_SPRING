## Random Numbers
- Case : roll a dice and climb stairs
    1. If 1, 2 : -1
    2. If 3, 4, 5 : +1
    3. If 6 : roll again, go up the result
    4. Can't go below 0
    5. 0.1% chance of 'falling down' -> start from 0 again
    6. End : reach 60

- NumPy `np.random.rand()`
    - `np.random.seed()` : same seed, same random numbers
    - `np.random.randint(start, end)` : end is **NOT** included


````python
# simple example
import numpy as np
np.random.seed(123)
np.random.rand()
````

````python
# dice
print(np.random.randint(1, 7))

# one more roll
print(np.random.randint(1, 7))
````

````python
# the climbing game
# starting step
step = 50

# dice
dice = np.random.randint(1, 7)

# control construct
if dice <= 2 :
    step = step - 1
elif dice <=5 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# print out dice and step
print(dice)
print(step)
````

## Random Walk
- Record of random steps
- Use of `append()`

````python
# coin flip random walk
np.random.seed(123)
tails = [0]
for x in range(10) :
    coin = np.random.randint(0, 2)
    tails.append(tails[x]+coin)
print(tails)
````

````python
np.random.seed(123)
# random walk of the dice game
# Initialize random_walk
random_walk = [0]

for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    if dice <= 2:
        step = max(0, step-1) # Does not go below 0
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

print(random_walk)
````

````python
# visualization
import matplotlib.pyplot as plt
plt.plot(random_walk)
````

````python
# adding the 0.1% chance of going to 0

# Simulate random walk 250 times
all_walks = []
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.001 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
````
